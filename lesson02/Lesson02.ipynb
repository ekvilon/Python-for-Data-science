```python
t = np.linspace(0, 10, 51)
f = np.cos(t)

plt.plot(t, f, color='green')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.axis([0.5, 9.5, -2.5, 2.5])
```

## Задание 3

```python
x = np.linspace(-3, 3)
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(ncols = 2, nrows = 2)
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3)
plot1, plot2, plot3, plot4 = ax.flatten()
plot1.plot(x, y1)
plot1.set_title('График y1')
plot1.set_xlim(-5, 5)
plot2.plot(x, y2)
plot1.set_title('График y2')
plot3.plot(x, y3)
plot1.set_title('График y3')
plot4.plot(x, y4)
plot1.set_title('График y4')
```

```python
plt.style.use('fivethirtyeight')
```

```python
cc = pd.read_csv('creditcard.csv')
```

```python
fig, ax = plt.subplots(ncols=2, nrows=1)
plot1, plot2 = ax.flatten()
cc[['Class']].value_counts().plot(kind='bar', ax=plot1)
cc[['Class']].value_counts().plot(kind='bar', logy=True, ax=plot2)
```

```python
plt.hist(cc[['V1']][cc['Class'] == 0], density=True, bins=20, alpha=0.5, color='grey')
plt.hist(cc[['V1']][cc['Class'] == 1], density=True, bins=20, alpha=0.5, color='red')
plt.legend(labels=["Class 0", "Class 1"])
plt.xlabel('Class')
```
